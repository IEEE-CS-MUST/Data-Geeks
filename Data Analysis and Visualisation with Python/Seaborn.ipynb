# Visualizing statistical relationships
#### Understanding the relationships between variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
a=sns.load_dataset("flights")
sns.relplot(x="month",y="passengers",data=a,hue="year",kind="line")
```

# Plotting with categorical data
#### Dividing into discrete groups

```python
sns.catplot(x="month",y="passengers",data=a,kind="violin")  #box,boxen,swarm,strip
```

# Visualizing the distribution of a dataset
#### univariate and bivariate distributions

```python
sns.displot(x="month",y="passengers",data=a,hue="year")
```

# Multiplot grids
#### Graphs are plotted side by side using the same scale and axes to aid comparison

```python
b=sns.PairGrid(a)
b.map(plt.scatter)
sns.pairplot(a)
```

# Plot aesthetics
#### used to make interface more user friendly and customizing themes with high level interfaces

```python
sns.set(style="dark")  #white,whitegrid,darkgrid
sns.relplot(x="month",y="passengers",data=a)
```

#### color_palette

```python
c=sns.color_palette(['red','blue'])
sns.palplot(c)
```

# Some other majorly used functions
#### boxplot

```python
sns.boxplot(x="month",y="passengers",data=a)
```

#### countplot

```python
sns.countplot(x="month",data=a)
```

#### barplot

```python
sns.barplot(x="month",y="passengers",data=a)
```
