# Matplotlib Pyplot and Markers

```python
#(17,12) (22,15) (26,34) (35,40)
```

```python
import matplotlib.pyplot as plt
xpoints = (17,22,26,35)
ypoints = (12,15,34,40)
```

```python
plt.plot(xpoints,ypoints,marker = '*',ms='15',mec='r',mfc='y')
plt.show()
```

# Matplotlib Line, Labels and Grid

```python
plt.plot(xpoints,ypoints,ls = 'dashed', color='#445566', lw='4')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('DATA')
plt.grid(axis = 'x')
plt.show()
```

# Matplotlib Subplot

```python
x = (0, 1, 2, 3)
y = (3, 8, 1, 10)
plt.subplot(1, 2, 1)  #(row,column,position)
plt.plot(x,y)
#plot 2:
x = (0, 1, 2, 3)
y = (10, 20, 30, 40)
plt.subplot(1,2, 2)
plt.plot(x,y)
plt.show()
```

# Matplotlib Scatter

```python
x = (25,27,28,31,32,45,38,67,45)
y = (20,39,66,21,30,55,14,69,30)
plt.scatter(x,y)
x = (15,17,18,17,12)
y = (56,22,44,18,11)
plt.scatter(x,y,color='hotpink')
plt.show()
```

# Matplotlib Bars

```python
x = ("P", "Q", "R", "S")
y = (13, 28, 11, 20)
plt.barh(x,y,height=0.2)
plt.show()
```

# Matplotlib Pie charts

```python
y =(23,89,22,56,67)
mypiechart = ("q","e","p","h","k")
values = (0.3,0,0,0.5,0)
plt.pie(y, labels=mypiechart, explode = values)
plt.show()
```
