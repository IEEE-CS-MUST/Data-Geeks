### Handling the missing data

```python
# Missing Data - Director, Cast, Country, date_added, rating, duration
# Director, Cast and Country has a lot of missing values so we can't just drop it
# Country, Director and Cast with No data inplace

netflix_df['country'].replace(np.nan, 'No data', inplace=True)
netflix_df['cast'].replace(np.nan, 'No data', inplace=True)
netflix_df['director'].replace(np.nan, 'No data', inplace=True)

netflix_df.dropna(inplace=True)
netflix_df.drop_duplicates(inplace=True)

netflix_df.isnull().sum()


netflix_df.info()
netflix_df
```

```python
# Date 
netflix_df['date_added']=pd.to_datetime(netflix_df['date_added']) 

netflix_df['month_name_added']=netflix_df['date_added'].dt.month_name()
netflix_df['month_added']=netflix_df['date_added'].dt.month
netflix_df['year_added']=netflix_df['date_added'].dt.year

netflix_df.head()
```

# Let's begin data analysis

```python
# TV shows and Movie data
shows_df = netflix_df[netflix_df['type']=='TV Show']
movies_df = netflix_df[netflix_df['type']=='Movie']

shows_df.head()

```

```python
shows_df.shape[0]
```

```python
category = ['TV Shows', 'Movies']
count = [shows_df.shape[0], movies_df.shape[0]]
fig = plt.figure(figsize=(10,7))
plt.pie(count, labels= category, autopct="%.2f")
plt.show()
```

```python
# Countries
# We have multiple country for same content- we need primary country
netflix_df['country'].head(20)
```

```python
netflix_df['country_f'] = netflix_df['country'].apply(lambda x: x.split(',')[0])
netflix_df['country_f'].head(10)
```

```python
# Split function
"RM, Jin, Suga".split(",")[0]
```

```python
netflix_df['country_f'].replace('United States', 'USA', inplace=True)
netflix_df['country_f'].replace('United Kingdom', 'UK', inplace=True)
netflix_df['country_f'].replace('South Korea', 'S. Korea', inplace=True)
netflix_df['country_f']
```

```python
country_count = netflix_df['country_f'].value_counts().reset_index().rename(columns ={"index":"country", "country_f":"count"})
country_count.drop(2, inplace = True)
top_countries = country_count[:10]
top_countries.head(10)
```

```python
fig = plt.figure(figsize=(17,7))
plt.subplot(1,2, 1)
sns.barplot(x=top_countries['country'], y=top_countries['count'])
plt.subplot(1,2,2)
plt.pie(top_countries['count'], labels=top_countries['country'], autopct="%.2f")
plt.show()
```

```python
netflix_df
```

```python
# Amount of content added as function of time
col = 'year_added'

total_count = netflix_df[col].value_counts().reset_index().rename(columns = {"index":"year", "year_added":"count"})
movie_count = movies_df[col].value_counts().reset_index().rename(columns = {"index":"year", "year_added":"count"})
shows_count = shows_df[col].value_counts().reset_index().rename(columns = {"index":"year", "year_added":"count"})

total_count
```

```python
fig = plt.figure(figsize=(15,5))
plt.xticks(np.arange(2008, 2022, 1 ))
sns.lineplot(data = total_count, x = 'year', y='count')
sns.lineplot(data = movie_count, x='year', y = 'count')
sns.lineplot(data = shows_count, x='year', y = 'count')
plt.legend(['Total', 'Movies', 'Shows'])
plt.show()
```

```python
# Content ratings
plt.figure(figsize=(12,9))
sns.countplot(x='rating', data=netflix_df, hue='type')
```

```python
# month in which most content was released
month_order=['January', 'February', 'March', 'April', 'May','June','July','August','September','October','November','December']

sns.countplot(x='month_name_added', data=netflix_df, order=month_order)
sns.set_style('whitegrid')
plt.xticks(rotation = 90)
```

```python
# Actor with most content
netflix_df['cast']
cast_df = netflix_df['cast'].str.split(',', expand=True)
cast_df
```

```python
actor = cast_df.melt(value_name = 'Actor')['Actor']
actor = actor.value_counts().reset_index().rename(columns = {'index':'Actor', 'Actor':'count'})
actor.drop(0, inplace=True)
actor
```

```python
top = actor[:10]
sns.catplot(x = 'count', y='Actor', data=top)
plt.xticks(np.arange(22, 40, 1), rotation=90)
plt.show()
```

```python
# Indian Movie directors with most movies
movies_ind = netflix_df[netflix_df['country_f']=='India']
movies_ind = movies_ind[movies_ind['type']=='Movie']
ind_directors = movies_ind['director'].value_counts().reset_index().rename(columns={'index':'director', 'director':'count'})
ind_directors.drop(0, inplace=True)

dir = ind_directors[:15]

plt.barh(dir['director'], dir['count'])
plt.xticks(np.arange(0,10,1))
```
